```python
def mp(n):
    a0 = list(n[0])
    a0 = list(map(int,list(a0)))
    a1 = int(n[1])
    b = list(map(int,list(a0)))
    b.sort(reverse = True) #b는 목표값
    opp = 0
    p = 0    
    cnt = 0
    to_change = []
    
    for i in range(len(a0)): # 바뀌어야 할 숫자들
        if a0[i] != b[i]:
            to_change.append(a0[i])
    print(a0)
    print(b)
    
    print(to_change)
    
    if a0 == b:
        for i in range(a1): # 최대값에서 남은 횟수만큼 가장 작은 단위만 바꿈
            a0[-1], a0[-2] = a0[-2], a0[-1]
            return a0
    
    while(opp < a1):

        q = 0
        tmp0 = 0
        tmp1 = []
        if a0[p] == b[p]:
            p += 1
            
        else :
            r = 0
            tmp0 = a0[p]
            tmp1.append(a0[p])
            a0[p] = b[p]
            while(q == 0):
                #print(b[p])
                if a0[-1-r] != b[p] :
                    r += 1
                else :
                    a0[-1-r] = tmp0
                    q += 1
                    opp += 1
        
        if a0 == b:
            for i in range(a1 - opp): # 최대값에서 남은 횟수만큼 가장 작은 단위만 바꿈
                a0[-1], a0[-2] = a0[-2], a0[-1]
                return a0
    
    return a0
```

```python
num = int(input())
for i in range(num):
    numbers = input().split()
    ans = mp(numbers)
    ans = list(map(str,ans))
    ans = ''.join(ans)
    print(f'#{i+1} {ans}')
```
